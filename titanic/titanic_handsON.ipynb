## Modeling using Scikit Learn
---

```python
df_train.describe(include='O').columns
```

```python
def parse_ticket(ticket_string):
    ticket_list = ticket_string.split(" ")
    if len(ticket_list)>1:
        return ticket_list[0]
    else:
        return "Numbered ticket"

df_train['Parsed_Ticket']=df_train['Ticket'].apply(parse_ticket)
```

```python
df_train.pivot_table(index='Parsed_Ticket', columns='Survived',values='Sex', aggfunc='count', margins=True)
```

```python
def parse_cabin(cabin_string):
    cabin_string = str(cabin_string)
    if cabin_string[0] == 'n':
        return 0
    else:
        return 1

df_train['Has_Cabin'] = df_train['Cabin'].apply(parse_cabin)
df_train.pivot_table(index='Has_Cabin', columns='Survived',values='Sex', aggfunc='count', margins=True)
```

```python
df_train.pivot_table(index='Embarked', columns='Survived',values='Sex', aggfunc='count', margins=True)
```

```python
df_train = pd.get_dummies(df_train, columns=["Embarked", "Sex"])
```

```python
df_train.loc[df_train['Age'].isna(),:]
```

```python
df_train['Age'].fillna(df_train['Age'].median(), inplace=True)
```

```python
df_train.columns
```

```python
X = df_train[['Pclass', 'Age', 'SibSp','Parch','Fare','Has_Cabin','Embarked_C','Embarked_Q', 'Embarked_S', 'Sex_female']]
y = df_train['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
classifier = LogisticRegression()
```

```python
classifier = classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
```

```python
classifier.score(X_test, y_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
